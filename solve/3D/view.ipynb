```python
import numpy as np
import h5py,os
import matplotlib.pyplot as plt
```

```python
f = h5py.File('phi.h5', 'r')
dset = f[list(f.keys())[0]]
data = np.zeros(dset.shape)
dset.read_direct(data, np.s_[:,:,:], np.s_[:,:,:])
f.close()
```

```python
print(data.shape)
print(data.sum())
```

```python
f,ax = plt.subplots(1,1)
f.set_size_inches(12,10)

im = ax.pcolormesh(data[32][:][:], cmap='turbo')

ax.axes.set_aspect('equal')
plt.colorbar(im, ax=ax)
```
